### Create a new column score_level
"High" if total_score ≥ 90 , "Medium" if 70 ≤ total_score < 90, "Low" if total_score < 70

```python
def func(total_score):
  if total_score >= 90:
    return "High"
  elif total_score < 90 and total_score >= 70:
    return "Medium"
  else:
    return "Low"
```

```python
df["score_level"] = df["total_score"].apply(func)
df
```

### Make a new column for study efficiency = total_score ÷ weekly_self_study_hours.

### Create a new column study_category
"Light" if weekly_self_study_hours < 10,
"Moderate" if 10 ≤ hours ≤ 20,
"Heavy" if hours > 20

```python
def categorize_study(hours):
    if hours < 10:
        return "Light"
    elif hours <= 20: # Since it's not < 10, we know it's >= 10
        return "Moderate"
    else:
        return "Heavy"

df["study_category"] = df["weekly_self_study_hours"].apply(categorize_study)
df
```

```python
# Define the conditions based on your hours thresholds
conditions = [
    df["weekly_self_study_hours"] < 10,
    (df["weekly_self_study_hours"] >= 10) & (df["weekly_self_study_hours"] <= 20)
]

# Define the matching category labels
choices = ["Light", "Moderate"]

# np.select applies choices based on conditions; the default handles hours > 20 ("Heavy")
df["study_category"] = np.select(conditions, choices, default="Heavy")
df
```
